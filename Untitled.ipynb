```python
# TASK 1 
# Raw data
raw_students = [
    {"name": "  ayesha SHARMA  ", "roll": "101", "marks_str": "88, 72, 95, 60, 78"},
    {"name": "ROHIT verma",       "roll": "102", "marks_str": "55, 68, 49, 72, 61"},
    {"name": "  Priya Nair  ",    "roll": "103", "marks_str": "91, 85, 88, 94, 79"},
    {"name": "karan MEHTA",       "roll": "104", "marks_str": "40, 55, 38, 62, 50"},
    {"name": " Sneha pillai ",    "roll": "105", "marks_str": "75, 80, 70, 68, 85"},
]
```

```python
#Cleaning data

cleaned_students = []

for student in raw_students:
    clean_name = student["name"].strip().title()
    clean_roll = int(student["roll"])
    clean_marks = [int(mark) for mark in student["marks_str"].split(", ")]
    
    name_words = clean_name.split()
    is_valid = all(word.isalpha() for word in name_words)
    validation_marker = "Valid name" if is_valid else "Invalid name"
    
    cleaned_students.append({
        "name": clean_name,
        "roll": clean_roll,
        "marks": clean_marks,
        "validation": validation_marker
    })
cleaned_students
```

```python
#Printing Profile card
for student in cleaned_students:
    print("================================")
    print(f"Student : {student['name']} [{student['validation']}]")
    print(f"Roll No : {student['roll']}")
    print(f"Marks   : {student['marks']}")
    print("================================")
    print()
```

```python
#Printing lowrcase & upppercase
print("--- String Case Variations for Roll No 103 ---")
for student in cleaned_students:
    if student["roll"] == 103:
        print(f"ALL CAPS : {student['name'].upper()}")
        print(f"lowercase: {student['name'].lower()}")
        break
```

```python
# TASK 2
#Initial Data
student_name = "Ayesha Sharma"
subjects = ["Math", "Physics", "CS", "English", "Chemistry"]
marks = [88, 72, 95, 60, 78]

print(f"Loaded data for {student_name}.")
```

```python
# Generating the Grade report
print(f"--- Grade Report for {student_name} ---")

for i in range(len(subjects)):
    subject = subjects[i]
    mark = marks[i]
    
    if mark >= 90:
        grade = "A+"
    elif mark >= 80:
        grade = "A"
    elif mark >= 70:
        grade = "B"
    elif mark >= 60:
        grade = "C"
    else:
        grade = "F"
        
    print(f"{subject:10} | Marks: {mark:3} | Grade: {grade}")
```

```python
#Calculating the statistics

print("--- Statistics ---")

total_marks = sum(marks)
average_marks = round(total_marks / len(marks), 2)

max_mark = max(marks)
min_mark = min(marks)

highest_subject = subjects[marks.index(max_mark)]
lowest_subject = subjects[marks.index(min_mark)]

print(f"Total Marks   : {total_marks}")
print(f"Average Marks : {average_marks}")
print(f"Highest Score : {highest_subject} ({max_mark})")
print(f"Lowest Score  : {lowest_subject} ({min_mark})")
```

```python
#Marks entry system
print("--- Interactive Marks Entry System ---")

new_subjects_added = 0

while True:
    new_subject = input("Enter subject name (or type 'done' to finish): ").strip()
    
    if new_subject.lower() == 'done':
        break
        
    marks_input = input(f"Enter marks for {new_subject} (0-100): ").strip()
    
    try:
        new_mark = float(marks_input)
        
        if new_mark < 0 or new_mark > 100:
            print("Do not crash, and do not add invalid entries to the list.\n")
            continue 
            
    except ValueError:
        print("Do not crash, and do not add invalid entries to the list\n")
        continue 
        
    subjects.append(new_subject)
    marks.append(new_mark)
    new_subjects_added += 1
    print(f"✓ Successfully added {new_subject} with {new_mark} marks.\n")

print("\n--- Final Update ---")
print(f"New subjects added: {new_subjects_added}")

if len(marks) > 0:
    updated_total = sum(marks)
    updated_average = round(updated_total / len(marks), 2)
    print(f"Updated Average across all {len(marks)} subjects: {updated_average}")
```

```python
#TASK 3:
#Dataset
class_data = [
    ("Ayesha Sharma",  [88, 72, 95, 60, 78]),
    ("Rohit Verma",    [55, 68, 49, 72, 61]),
    ("Priya Nair",     [91, 85, 88, 94, 79]),
    ("Karan Mehta",    [40, 55, 38, 62, 50]),
    ("Sneha Pillai",   [75, 80, 70, 68, 85]),
]

print("✓ Class data loaded successfully.")
```

```python
# Initialize tracking variables
passed_count = 0
failed_count = 0
topper_name = ""
topper_average = -1  
sum_of_all_averages = 0

print(f"{'Name':<18} | {'Average':<7} | Status")
print("-" * 40)

# Processing each student
for name, marks in class_data:
    average = sum(marks) / len(marks)
    
    if average >= 60:
        status = "Pass"
        passed_count += 1
    else:
        status = "Fail"
        failed_count += 1
        
    if average > topper_average:
        topper_average = average
        topper_name = name
        
    sum_of_all_averages += average
    
    print(f"{name:<18} | {average:>7.2f} | {status}")
```

```python
# Calculating final class average
class_average = sum_of_all_averages / len(class_data)
print("--- Class Summary ---")
print(f"Passed       : {passed_count}")
print(f"Failed       : {failed_count}")
print(f"Class Topper : {topper_name} ({topper_average:.2f})")
print(f"Class Average: {class_average:.2f}")
```
